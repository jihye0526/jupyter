## 데이터에 함수 적용 (apply)

```python
# 나이 뒤에 '살'을 붙이는 역할
def add_age(age):
    return str(age) + '살'

df['나이'] = df['나이'].apply(add_age) # 나이 데이터에 대해서 add_age 함수를 호출한 결과 데이터를 반영
df
```

```python
def capitalize(name):
    if pd.notnull(name): # NaN 이 아닌 경우
        return name.capitalize() # 첫 글자는 대문자로, 나머지는 소문자로
    return name

df['활동명'] = df['활동명'].apply(capitalize)
df
```
