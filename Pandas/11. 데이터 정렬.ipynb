# 11. 데이터 정렬

```python
import pandas as pd
df = pd.read_excel('seventeen.xlsx', index_col='지원번호')
df
```

```python
df.sort_values('생일(월)') # 생일(월) 기준으로 오름차순 정렬
```

```python
df.sort_values('생일(월)', ascending=False) # 생일(월) 기준으로 내림차순 정렬
```

```python
df.sort_values(['나이', '생일(월)'], ascending=False) # 나이, 생일(월) 기준으로 내림차순
```

```python
df.sort_values(['나이', '생일(월)']) # 나이, 생일(월) 기준으로 오름차순
```

```python
df.sort_values(['나이', '생일(월)'], ascending=[False, True]) # 나이는 내림차순, 생일(월)은 오름차순 정렬
```

```python
df['이름'].sort_values()
```

```python
df['이름'].sort_values(ascending=False)
```

```python
# df.sort_index(), index가 문자열이므로 정렬을 하면 10~13 출력 후 1~9가 나옴
a = df.index.to_series().str.rsplit('번').str[0].apply(pd.to_numeric).sort_values()

df = df.reindex(index=a.index)
df
```

```python
# df.sort_index(ascending=False) # index가 문자열이므로 문자인 '번'을 잘라 숫자형으로 변환 후 정렬
a = df.index.to_series().str.rsplit('번').str[0].apply(pd.to_numeric).sort_values(ascending = False)

df = df.reindex(index=a.index)
df
```
