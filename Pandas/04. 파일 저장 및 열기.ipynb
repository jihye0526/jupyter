# 4. 파일 저장 및 열기
DataFrame 객체를 excel, csv, txt 등 형태의 파일로 저장 및 열기

```python
import pandas as pd

data = {
     '이름' : ['최승철', '윤정한' ,'홍지수', '문준휘', '전원우', '권순영', '이지훈', '서명호', '김민규', '이석민', '부승관', '최한솔', '이찬'],
     '유닛' : ['힙합팀', '보컬팀', '보컬팀', '퍼포팀', '힙합팀', '퍼포팀', '보컬팀', '퍼포팀', '힙합팀', '보컬팀', '보컬팀', '힙합팀', '퍼포팀'],
     '나이' : [27, 27, 27, 26, 26, 26, 26, 25, 25, 25, 24, 24, 23],
     '생일(월)' : [8, 10, 12 ,6, 7, 6, 11, 11, 4, 2, 1, 2, 2]
}

df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번', '9번', '10번', '11번', '12번', '13번'])
df.index.name = '지원번호'
df
```

## 저장하기
### CSV 파일로 저장

```python
df.to_csv('seventeen.csv')
```

```python
df.to_csv('seventeen.csv', encoding='utf-8-sig', index=False)
```

### 텍스트(.txt) 파일로 저장

```python
df.to_csv('seventeen.txt', sep='\t') # tab으로 구분된 텍스트 파일
```

### 엑셀 파일로 저장

```python
df.to_excel('seventeen.xlsx')
```

---
## 열기
### csv 파일 열기

```python
df = pd.read_csv('seventeen.csv')
df
```

```python
df = pd.read_csv('seventeen.csv', skiprows=1) # 지정된 갯수 만큼의 row를 건너뜀
df
```

```python
df = pd.read_csv('seventeen.csv', skiprows=[1, 3, 5])  # 1, 3, 5 row는 제외 (row는 0부터 시작)
df
```

```python
df = pd.read_csv('seventeen.csv', nrows=4) # 지정된 갯수 만큼의 row만 가져옴
df
```

```python
df = pd.read_csv('seventeen.csv', skiprows=2, nrows=4) # 처음 2 row 무시, 이후에 4 row 가져옴
df
```

### 텍스트(.txt) 파일 열기

```python
df = pd.read_csv('seventeen.txt', sep='\t')
df
```

```python
df = pd.read_csv('seventeen.txt', sep='\t', index_col='지원번호')
df
```

```python
df = pd.read_csv('seventeen.txt', sep='\t')
df.set_index('지원번호', inplace=True)
df
```

### 엑셀(.xlsx) 파일 열기

```python
df = pd.read_excel('seventeen.xlsx')
df
```

```python
df = pd.read_excel('seventeen.xlsx', index_col='지원번호')
df
```
