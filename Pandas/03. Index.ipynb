# 3. Index
데이터에 접근할 수 있는 주소 값

```python
import pandas as pd

data = {
     '이름' : ['최승철', '윤정한' ,'홍지수', '문준휘', '전원우', '권순영', '이지훈', '서명호', '김민규', '이석민', '부승관', '최한솔', '이찬'],
     '유닛' : ['힙합팀', '보컬팀', '보컬팀', '퍼포팀', '힙합팀', '퍼포팀', '보컬팀', '퍼포팀', '힙합팀', '보컬팀', '보컬팀', '힙합팀', '퍼포팀'],
     '나이' : [27, 27, 27, 26, 26, 26, 26, 25, 25, 25, 24, 24, 23],
     '생일(월)' : [8, 10, 12 ,6, 7, 6, 11, 11, 4, 2, 1, 2, 2]
}

df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번', '9번', '10번', '11번', '12번', '13번'])
df
```

```python
df.index
```

## Index 이름 설정

```python
df.index.name = '지원번호'
df
```

## Index 초기화

```python
df.reset_index()
```

```python
df.reset_index(drop=True) # 원래 쓰던 '지원번호' 인덱스 삭제
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True) # 실제 데이터에 바로 반영
```

```python
df
```

## Index 설정
지정한 column으로 Index 를 설정

```python
df.set_index('이름')
```

```python
df
```

```python
df.set_index('이름', inplace=True)
```

```python
df
```

## Index 정렬
index 를 기준으로 오름차순, 내림차순 정렬

```python
df.sort_index() # 인덱스를 오름차순 정렬
```

```python
df.sort_index(ascending=False) # 내림차순 정렬
```
