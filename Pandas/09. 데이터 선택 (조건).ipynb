# 9. 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('seventeen.xlsx', index_col='지원번호')
df
```

```python
df['나이'] >= 26 # 지원자들의 나이가 26 이상인지 여부를 True / False
```

```python
filt = (df['나이'] >= 26)
df[filt]
```

```python
df[~filt] # filter 를 역으로 적용
```

```python
df[df['나이'] >= 26]
```

```python
df.loc[df['나이'] >= 26, '이름'] # 나이가 26 이상인 지원자들의 이름 데이터
```

```python
df.loc[df['나이'] >= 26, ['이름', '생일(월)']]
```

## 다양한 조건

### & 그리고

```python
df.loc[(df['나이'] >= 26) & (df['유닛'] == '힙합팀')] # 나이가 26 이상인 지원자들 중 힙합팀 데이터
```

### | 또는

```python
df.loc[(df['나이'] >= 27) | (df['나이']<=23)] # 나이가 27 이상이거나, 23 이하인 지원자의 데이터
```

### str 함수

```python
filt = df['이름'].str.startswith('이') # '이'씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('지') # 이름에 '지'가 들어가는 사람
df[filt]
```

```python
df[~filt] # 이름에 '지'가 들어가지 않는 사람
```

```python
units = ['보컬팀', '퍼포팀']
filt = df['유닛'].isin(units) # 유닛이 보컬팀이거나 퍼포팀인 사람
df[filt]
```
