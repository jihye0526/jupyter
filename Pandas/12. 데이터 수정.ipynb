# 12. 데이터 수정

```python
import pandas as pd

df = pd.read_excel('seventeen.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
df['나이'].replace({27 : 28, 26 : 27, 25 : 26, 24 : 25, 23 : 24})
```

```python
df
```

```python
df['나이'].replace({27 : 28}, inplace=True)
```

```python
df
```

```python
df['MBTI'] = df['MBTI'].str.lower()
df
```

```python
df['유닛2'].fillna('준비 중', inplace=True)
```

```python
df
```

```python
df['MBTI'] = df['MBTI'].str.upper()
df
```

```python
df['이름'] = df['이름'] + '()'
df
```
