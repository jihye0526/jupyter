# 14. 그룹화
동일한 값을 가진 것들끼리 합쳐서 통계 또는 평균 등의 값을 계산하기 위해 사용

```python
import pandas as pd

df = pd.read_excel('seventeen.xlsx', index_col='지원번호')
df
```

```python
df.groupby('유닛')
```

```python
df.groupby('유닛').get_group('힙합팀')
```

```python
df.groupby('유닛').get_group('보컬팀')
```

```python
df.groupby('유닛').mean() # 계산 가능한 데이터들의 평균값
```

```python
df.groupby('유닛').size() # 각 그룹의 크기
```

```python
df.groupby('유닛').size()['보컬팀'] # 유닛으로 그룹화를 한 뒤에 보컬팀에 해당하는 데이터의 수
```

```python
df.groupby('유닛')['나이'].mean() # 유닛별로 그룹화를 한 뒤에 나이의 평균 데이터
```

```python
df.groupby('유닛')[['나이', '생일(월)']].mean() # 유닛으로 그룹화를 한 뒤에 나이, 생일(월) 평균 데이터
```

```python
# df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2] # 학년 Column 추가
# df
```

```python
df.groupby(['유닛', '생일(월)']).mean() # 유닛별, 생일(월)별 평균 데이터
```

```python
df.groupby('생일(월)').mean() # 생일(월)별 평균 데이터
```

```python
df.groupby('유닛').mean().sort_values('나이', ascending=False)
```

```python
df.groupby(['유닛', '나이']).sum()
```

```python
df.groupby('유닛')[['이름', '유닛2']].count() # 유닛으로 그룹화를 한 뒤에 각 유닛별 이름, 유닛2를 가지고 있는 데이터의 수를 가져옴
```

```python
unit = df.groupby('유닛')
unit['유닛2'].value_counts() # 유닛으로 그룹화를 한 뒤에 유닛2별 지원자 수를 가져옴
```

```python
unit['유닛2'].value_counts().loc['보컬팀'] # 유닛으로 그룹화 한 뒤에 보컬팀에 대해서 유닛2별 지원자 수를 가져옴
```

```python
unit['유닛2'].value_counts().loc['퍼포팀'] # 유닛으로 그룹화 한 뒤에 퍼포팀에 대해서 유닛2별 지원자 수를 가져옴
```

```python
unit['유닛2'].value_counts(normalize=True).loc['보컬팀'] # 유닛으로 그룹화를 한 뒤 보컬팀에 대해 유닛2별로 멤버를 퍼센트로 나타냄
```
