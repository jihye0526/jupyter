## Data 준비
사전 (dict) 자료구조를 통해 생성

예) 세븐틴 멤버에 대한 데이터

```python
data = {
     '이름' : ['최승철', '윤정한' ,'홍지수', '문준휘', '전원우', '권순영', '이지훈', '서명호', '김민규', '이석민', '부승관', '최한솔', '이찬'],
     '유닛' : ['힙합팀', '보컬팀', '보컬팀', '퍼포팀', '힙합팀', '퍼포팀', '보컬팀', '퍼포팀', '힙합팀', '보컬팀', '보컬팀', '힙합팀', '퍼포팀'],
     '나이' : [27, 27, 27, 26, 26, 26, 26, 25, 25, 25, 24, 24, 23],
     '생일(월)' : [8, 10, 12 ,6, 7, 6, 11, 11, 4, 2, 1, 2, 2]
}
data
```

```python
data['이름']
```

```python
data['생일(월)']
```

## DataFrame 객체 생성

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df
```

## 데이터 접근

```python
df['이름']
```

```python
df['유닛']
```

```python
df[['이름', '유닛']]
```

## DataFrame 객체 생성 (Index 지정)

```python
df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번', '9번', '10번', '11번', '12번', '13번'])
df
```

## DataFrame 객체 생성 (Column 지정)
data 중에서 원하는 column 만 선택하거나, 순서 변경 가능

```python
df = pd.DataFrame(data, columns=['이름', '유닛', '생일(월)'])
df
```
