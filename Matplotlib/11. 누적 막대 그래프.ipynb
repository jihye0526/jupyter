# 11. 누적 막대 그래프

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 깨짐 현상을 해결
```

```python
df = pd.read_excel('../Pandas/seventeen.xlsx')
df
```

```python
plt.bar(df['이름'], df['나이'])
plt.bar(df['이름'], df['생일(월)'], bottom=df['나이'])

plt.xticks(rotation=45)
```

```python
plt.bar(df['이름'], df['나이'], label='나이')
plt.bar(df['이름'], df['생일(월)'], bottom=df['나이'], label='생일(월)')

plt.xticks(rotation=45)
plt.legend()
```
