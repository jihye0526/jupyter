# 8. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 깨짐 현상을 해결
```

```python
labels = ['최승철', '권순영', '이지훈'] # 이름
values = [27, 26, 26] # 나이

plt.bar(labels, values)
```

```python
labels = ['최승철', '권순영', '이지훈'] # 이름
values = [27, 26, 26] # 나이

plt.bar(labels, values, color='salmon')
```

```python
labels = ['최승철', '권순영', '이지훈'] # 이름
values = [27, 26, 26] # 나이
colors = ['red', 'orange', 'salmon']

plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
labels = ['최승철', '권순영', '이지훈'] # 이름
values = [27, 26, 26] # 나이

plt.bar(labels, values)
plt.ylim(23, 28)
```

```python
plt.bar(labels, values, width=0.5)
```

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45) # x 축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45) # y 축의 나이 데이터 각도를 45도로 설정
```

```python
labels = ['최승철', '권순영', '이지훈'] # 이름
values = [27, 26, 26] # 나이
ticks=['힙합팀', '퍼포먼스팀', '보컬팀'] # 소속유닛

plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=90)
```
