# 9. 막대 그래프 (심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 깨짐 현상을 해결
```

```python
labels = ['최승철', '권순영', '이지훈'] # 이름
values = [27, 26, 26] # 나이

plt.barh(labels, values)
plt.xlim(24, 28)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/') # ////
bar[1].set_hatch('x') # xxxx
bar[2].set_hatch('..') #...
```

```python
bar = plt.bar(labels, values)
plt.ylim(24, 28)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.1, values[idx], ha='center', color='b')
```
