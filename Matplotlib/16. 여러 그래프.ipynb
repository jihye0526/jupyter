# 16. 여러 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 깨짐 현상을 해결
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/seventeen.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20, 15)) # 2 x 2 에 해당하는 plot 들을 생성
fig.suptitle('여러 그래프 넣기')

# 첫 번째 그래프
axs[0, 0].bar(df['이름'], df['나이'], label='나이') # 데이터 설정
axs[0, 0].set_title('첫 번째 그래프') # 제목
axs[0, 0].legend() # 범례
axs[0, 0].set(xlabel='이름', ylabel='나이') # x, y 축 label
axs[0, 0].set_facecolor('lightyellow') # 전경색
axs[0, 0].grid(linestyle='--', linewidth=0.5)
axs[0, 0].tick_params(rotation=45)

# 두 번째 그래프
axs[0, 1].plot(df['이름'], df['생일(월)'], label='생일(월)')
axs[0, 1].plot(df['이름'], df['생일(일)'], label='생일(일)')
axs[0, 1].tick_params(rotation=45)
axs[0, 1].legend()

# 세 번째 그래프
axs[1, 0].barh(df['이름'], df['나이'])

# 네 번째 그래프
axs[1, 1].plot(df['이름'], df['생일(월)'], color='green', alpha=0.5)
axs[1, 1].tick_params(rotation=45)
```
