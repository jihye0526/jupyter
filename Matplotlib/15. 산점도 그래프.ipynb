```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 깨짐 현상을 해결
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/seventeen.xlsx')
df
```

```python
df['가족 수'] = [4, 4, 3, 4, 4, 4, 3, 3, 4, 4, 5, 4, 4]
df
```

```python
plt.scatter(df['생일(월)'],df['생일(일)'])
plt.xlabel('생일(월)')
plt.ylabel('생일(일)')
```

```python
import numpy as np
sizes = np.random.rand(13) * 1000
sizes
```

```python
plt.scatter(df['생일(월)'],df['생일(일)'], s=sizes)
plt.xlabel('생일(월)')
plt.ylabel('생일(일)')
```

```python
sizes = df['가족 수'] * 200 
plt.scatter(df['생일(월)'],df['생일(일)'], s=sizes)
plt.xlabel('생일(월)')
plt.ylabel('생일(일)')
```

```python
sizes = df['가족 수'] * 200 
plt.scatter(df['생일(월)'],df['생일(일)'], s=sizes, c=df['가족 수'], cmap='viridis')
plt.xlabel('생일(월)')
plt.ylabel('생일(일)')
```


https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
sizes = df['가족 수'] * 200 
plt.scatter(df['생일(월)'],df['생일(일)'], s=sizes, c=df['가족 수'], cmap='viridis', alpha=0.3)
plt.xlabel('생일(월)')
plt.ylabel('생일(일)')
```

```python
plt.figure(figsize=(10, 10))

plt.scatter(df['생일(월)'],df['생일(일)'], s=sizes, c=df['가족 수'], cmap='viridis', alpha=0.3)
plt.xlabel('생일(월)')
plt.ylabel('생일(일)')

plt.colorbar(ticks=[3,4,5], label='가족 수', shrink=0.5, orientation='horizontal')
```
